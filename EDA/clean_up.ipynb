## Dependencies

```python
# Dependencies and Setup
import pandas as pd
import numpy as np
```

## Load and Check Data

```python
colnames= ['obj_ID','alpha','delta','u','g','r','i','z','run_ID','rerun_ID', 'cam_col','field_ID','spec_obj_ID','class', 'redshift', 'plate', 'MJD', 'fiber_ID']
```

```python
#Read in csv as dataframe
d = pd.read_csv("../Resources/top_200k.csv",names=colnames, header=None,low_memory=False)
data= pd.DataFrame(data=d)
data.head()
```

```python
#Read in QSO csv as dataframe
q = pd.read_csv("../Resources/qso_t_40k_z.csv",names=colnames, header=None,low_memory=False)
qso= pd.DataFrame(data=q)
qso.head()
```

```python
#Drop first two rows
data = data.drop([0, 1])
qso = qso.drop([0])
```

```python
frames = [data,qso]
df = pd.concat(frames)
df.head()
```

```python
#(Rows, Columns)
df.shape
```

```python
# Amount of Non-nulls and dtypes per column
df.info()
```

```python
df = df.reindex(columns=colnames)
```

```python
df['obj_ID'] = pd.to_numeric(df['obj_ID'], errors='coerce')
df['alpha'] = pd.to_numeric(df['alpha'], errors='coerce')
df['delta'] = pd.to_numeric(df['delta'], errors='coerce')
df['u'] = pd.to_numeric(df['u'], errors='coerce')
df['g'] = pd.to_numeric(df['g'], errors='coerce')
df['r'] = pd.to_numeric(df['r'], errors='coerce')
df['i'] = pd.to_numeric(df['i'], errors='coerce')
df['z'] = pd.to_numeric(df['z'], errors='coerce')
df['run_ID'] = pd.to_numeric(df['run_ID'], downcast='signed', errors='coerce')
df['rerun_ID'] = pd.to_numeric(df['rerun_ID'], errors='coerce')
df['cam_col'] = pd.to_numeric(df['cam_col'], errors='coerce')
df['field_ID'] = pd.to_numeric(df['field_ID'], errors='coerce')
df['spec_obj_ID'] = pd.to_numeric(df['spec_obj_ID'], errors='coerce')
df['redshift'] = pd.to_numeric(df['redshift'], errors='coerce')
df['plate'] = pd.to_numeric(df['plate'], errors='coerce')
df['MJD'] = pd.to_numeric(df['MJD'], errors='coerce')
df['fiber_ID'] = pd.to_numeric(df['fiber_ID'], errors='coerce')
```

```python
df.info()
```

```python
#Amount of unique entries per column
df.nunique()
```

```python
#Check for duplicate data
duplicate = df.duplicated(["spec_obj_ID"])
print(duplicate.sum())
df[duplicate]
```

```python
#Drop duplicate rows
df = df.drop_duplicates(subset=['spec_obj_ID'])
```

```python
#Analysis on stars
df_star = df.loc[(df['class'] == 'STAR')]
```

```python
#Analysis on Galaxies
df_galaxy= df.loc[(df['class'] == 'GALAXY')]
```

```python
#Analysis of quasars
df_quasar = df.loc[(df['class'] == 'QSO')]
```

```python
#Count number of quasars
df_quasar.count()
```

```python
# #To csv file 
df.to_csv("data.csv",index=False)
```
